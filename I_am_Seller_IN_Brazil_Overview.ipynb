### 판매 - 주문 지역 간 거래 횟수 비교 

````python
df1 = df_location[['seller_city','customer_city']].copy()

# 판매지역과 주문지역의 하나의 쌍 tuple 생성 
df1['seller_customer_pair'] = df1.apply(lambda x: (x['seller_city'], x['customer_city']), axis = 1)

# value_count 개수 저장 
seller_single = df1[['seller_city']].apply(pd.value_counts).sum(1) 
customer_single = df1[['customer_city']].apply(pd.value_counts).sum(1)

# 주문지역별, 판매지역별 개수
df1 = df1.assign(seller_count = df1['seller_city'].map(seller_single))
df1 = df1.assign(customer_count = df1['seller_city'].map(customer_single))
df1['single_count'] = df1['seller_count'] + df1['customer_count']

pair = df1[['seller_customer_pair']].apply(pd.value_counts).sum(1)
df1 = df1.assign(pair_count = df1['seller_customer_pair'].map(pair))

df1 = df1.drop_duplicates('seller_customer_pair').reset_index(drop = True)
df1[['seller_city','customer_city','seller_customer_pair','pair_count']].head(5)
````

````python
counts = 0
for i in range(df1.shape[0]):
    counts = counts + df1.loc[i,'pair_count']

df2 = df1.drop_duplicates('seller_city')
df2 = df2.sort_values(by = 'single_count', ascending= False).reset_index(drop = True)

print('총 주문 건수 : ', counts)
print('모든 seller & customer_cities의 유니크한 조합 개수 :', df2.shape[0])
df2.head()
````

### 위 경도 좌표 사이 거리 구하기 


- 위도와 경도 선 사이의 거리가 있기 때문에 지리학에는 적용되지 않는다.   
- 적도에 가까워 질수록 위도선은 더 멀어진다. 
- 지구의 곡률을 고려하여 거리를 측정해야한다. 

- haversine distance
image.png

````python
from math import radians , cos, sin, asin, sqrt 
from tqdm import tqdm
def haversine_distance(x1, y1, x2, y2 ):
    x1, y1, x2, y2 = map(radians , [x1,y1,x2,y2])
    d_lng = x2 - x1
    d_lat = y2 - y1
    a = sin(d_lat/2)**2 + cos(y1) * cos(y2) * sin(d_lng/2)**2
    dist = round((2*asin(sqrt(a))*6371*1000)/1000,3)
    return dist 
````

##### **Monthly Report : Category Cluster & Seller Z-Scores**

````python
def evaluation_table(df):
    res = df.groupby('product_category_name').agg({"order_id":"count", 
                                                   "product_id":'nunique', 
                                                   "seller_id":'nunique', 
                                                   'price':'sum'})
    res['price_mean'] = round(res['price']/res['seller_id'], 2)
    res['order_mean'] = round(res['order_id']/res['seller_id'], 2)
    res['price_std'] = 0
    res['order_std'] = 0
    res = res.reset_index()

    a = df.groupby(['seller_id', 'product_category_name']).agg({"order_id":"count",
                                                                "product_id":'nunique', 
                                                                'price':'sum'}).reset_index()

    # 표준편차
    for i in range(len(res)):
        # price_std
        d = a.loc[a['product_category_name'] == res.iloc[i, 0], 'price'] - res.iloc[i, 5]
        s = round((sum(d**2)/len(d))**1/2, 2)
        res.iloc[i, 7] = s

        # order_std
        d = a.loc[a['product_category_name'] == res.iloc[i, 0], 'order_id'] - res.iloc[i, 6]
        s = round((sum(d**2)/len(d))**1/2, 2)
        res.iloc[i, 8] = s
    
    return res
````

````python
cols = ['order_id', 'product_id', 'seller_id', 'price', 'product_category_name']

df_year_list = [] 

for year in ['2018-01','2018-02','2018-03','2018-04','2018-05','2018-06','2018-07','2018-08']:
    df_year_list.append(df.loc[df['year_month'] == year, cols])

for i in df_year_list:
    print(i.shape, end=' ')
````

````python
df['year_month'].value_counts(1).index.sort_values()
````

````python
evaluation_table_lst = [evaluation_table(i) for i in df_year_list]
evaluation_table_dict = dict(zip([ '2018-01', '2018-02', '2018-03',
       '2018-04', '2018-05', '2018-06', '2018-07', '2018-08', '2018-09'], evaluation_table_lst))
````

````python
def market_cluster(df):
    if len(df) < 4:
        return 0
    # 판매자의 등급을 군집화 
    scaler = StandardScaler()

    # K-Means Clustering 
    model = KMeans(n_clusters=4, random_state=42)

    pipeline = make_pipeline(scaler, model)
    pipeline.fit(df[['order_id', 'price']])

    market = df[['order_id', 'price']].copy()
    market["cluster"] = pipeline.predict(df[['order_id', 'price']])
    market['product_category_name'] = df['product_category_name']  
    return market

def cluster_tier(df, zero, one, two, three):
    df['cluster'] = df['cluster'].replace(0, zero)
    df['cluster'] = df['cluster'].replace(1, one)
    df['cluster'] = df['cluster'].replace(2, two)
    df['cluster'] = df['cluster'].replace(3, three)
    
    plt.figure(figsize=(6, 6))
    sns.scatterplot(data=df, x='order_id', y='price', hue='cluster', palette="Set1")

    return df
````

````python
market_tier_lst = [market_cluster(i) for i in evaluation_table_lst]
market_tier_dict = dict(zip(['2018-01', '2018-02', '2018-03',
       '2018-04', '2018-05', '2018-06', '2018-07', '2018-08', '2018-09'], market_tier_lst))
````

##### **주문건수를 기준으로 상관성을 보이는 품목 조합이 있는가?**

A의 시장에 진출을 원하지만 A 가 B 와 주문량이 높은 상관계수를 보인다면 B도 함께 판매하면 좋을 것 같다는 생각이 들었다. 

- 주문량 `Top 500 seller` 는 평균 4개의 품목을 취급하고 있고 주문량 `Bottom 500 seller` 는 평균 1개의 품목을 취급하고 있다는 결과를 참고하여 품목별 주문건수와 판매량이 정비례관계를 나타내는 품목들을 찾아보고자 한다.

````python
# 파생 변수 생성 : top15 list에 포함된 품목이면 True 아니면 False
df["top15"] = df["product_category_name_english"].map(lambda x: True if x in top15 else False)

# 파생 변수가 True인 것만 추출 = top15 list에 포함된 품목 데이터만 추출
condition = df["top15"] == True
df_top15 = df.loc[condition, :]

# pivot_table : 행=날짜, 열=품목, 값=품목별 날짜별 주문 건수 
temp_corr = pd.pivot_table(data=df_top15, index="year_month", columns="product_category_name_english", values="order_id", aggfunc="count")

# pivot_table 상관계수
top15_corr = temp_corr.corr()

# 히트맵 파라미터 mask 설정
mask = np.triu(np.ones(top15_corr.shape))

# 시각화
# 그래프 사이즈
plt.figure(figsize=(20, 15))
# 폰트 사이즈
plt.rc('font', size=15)
# 히트맵 시각화
sns.heatmap(top15_corr, cmap="coolwarm", mask=mask, annot=True, 
            vmin=0, vmax=1, annot_kws={"size":15}).set_title("품목별 주문 건수 상관계수", fontsize=25)
````

그러나 양의 상관성을 가진다고 하더라도, 이 제품들 쌍에 대해 어떠한 연관성이 있었는지, 어떤 기준으로 판매자가 조합을 구성하였는지, 실제로 A와 B가 묶여서 판매가 됐었는지는 알 수 없다.

따라서, 우리는 장바구니 분석을 통해 판매자 관점에서 어떤 클러스터를 기준으로 판매 물품 종류를 정했는지, 어떤 조합의 품목들이 가장 판매 빈도 수가 높았는지를 파악한 다음, 소비자의 관점에서 접근하여 이 조합이 실제로도 소비자들에게 선호 되었던 조합이었는지 연관성을 알아보기로 하였다.

##### **장바구니 분석을 통한 상품 간 연관성 분석**

앞서 주문 건수가 서로 비례하는 품목들이 있었는데, 여기서 각 판매자들이 어떤 종류의 품목들을 함께 판매하고 있는지를 알아보고자 장바구니 분석을 진행하였다. 

먼저, 소비자의 주문 정보 데이터를 기준으로 유망한 가치를 가지는 상품들의 조합을 찾는다. 

어떤 기준으로 판매 상품을 추천해줄 수 있을지 연관성 분석 결과 지표를 활용하여 설명하고자 하였지만,  소비자의 경우 동일한 한 명의 사람이 다른 물건을 함께 구매한 경우가 극히 드물다는 결과가 도출되어 유의미한 연관성을 찾을 수 없었다.



````python
seller_product = df_seller_customer.groupby('seller_id')['product_category_name'].unique()
customer_product = df_seller_customer.groupby('customer_id')['product_category_name'].unique()

product_by_seller = [] 
for i in seller_product.values.tolist():
    product_by_seller.append(i.tolist())

product_by_customer = [] 
for i in customer_product.values.tolist():
    product_by_customer.append(i.tolist())


# 연관분석
word_network = (list(apr(product_by_seller, min_support=0.01)))
df_nerwork=pd.DataFrame(word_network)
df_nerwork['length'] = df_nerwork['items'].apply(lambda x: len(x))

# length가 2인 row만 가져오기 
# 타이틀 내 동시출현하는 2개 단어들을 의미 
df_nerwork.shape
````

##### **Seller - 한 셀러가 여러 개의 품목을 팔 때 어떤 종류의 상품을 추가로 판매하는가 ?**

````python
# seller 기준 장바구니 
te = TransactionEncoder()
te_result = te.fit(product_by_seller).transform(product_by_seller, sparse=True)
te_df = pd.DataFrame.sparse.from_spmatrix(te_result, columns=te.columns_)
````

````python
itemset = apriori(te_df,
                           min_support=0.01, 
                           max_len=3, 
                           use_colnames=True, 
                          )
itemset.sort_values(by='support', ascending=False)
````

````python
metrics = association_rules(itemset, metric="confidence", min_threshold=0.01) 
metrics.sort_values(by='lift', ascending =False)
````

lift가 크면 클 수록 두 상품 사이의 연관성이 어느 정도 있어 보인다 라고 유추할 수 있다. 

- lift가 1이면 두 품목은 독립이지만 상위 row들은 향상도가 1보다 크기 때문에 양의 상관성을 가지며 이 것은 우연적 기회보다 높은 확률로 두 상품이 함께 가치를 가진다 라고 유추해볼 수 있다.
- 따라서 x를 구입할 때 z를 함께 판매할 것을 제안해볼 수 있다. 

##### **품목 동시 출현 빈도 네트워크망 시각화**

````python
# 노드의 크기는 해당 단어의 연결 중심성 크기를 나타낸 것이며 
# 해해당 단어가 얼마나 많이 출현했는지, 즉 얼마나 중요한지를 의미 

net = df_nerwork[(df_nerwork['length']==2) &(df_nerwork['support'] >=0.02)].sort_values(by='support', ascending =False)

G = nx.Graph()
ar = (net['items']); G.add_edges_from(ar)

#노드 순위 지정 위해 페이지랭크 계산
pr=nx.pagerank(G)
nsize = np.array([v for v in pr.values()])
nsize = 2000 * (nsize - min(nsize)) / (max(nsize) - min(nsize))
               
pos = nx.kamada_kawai_layout(G)
                  
plt.figure(figsize=(16,12)); plt.axis('off');
nx.draw_networkx(G, font_family='NanumBarunGothic', font_size=15,
                 pos=pos, node_color=list(pr.values()),node_size=nsize,
                 alpha=0.5, cmap=plt.cm.Blues);

plt.title('Product by Seller',fontsize=20);
plt.show();
````

- 각 노드의 크기는 각 판매 상품 조합의 **연결 중심성의 크기**를 나타낸 것이며 이는 해당 상품이 **얼마나 많이 출현**했는지, 즉 얼마나 중요한지를 의미한다.
    - 많이 등장한 상품일 수록 비즈니스적 가치가 높다는 의미를 가질 수 있다.
- 위 시각화 자료를 살펴보면,
    - Housewares 카테고리 상품을 판매하고 있는 사람은 Health_beauty, Auto, Sports_lessure 등의 카테고리 상품도 함께 판매하는 경우가 많았다고 볼 수 있다 .
    - Cool_stuff 는 Sports_lessure, Toys , Computes_accessories 등의 품목과 주로 판매 조합이 구성된다는 것을 알 수 있었다.
    - toys 카테고리의 경우 baby 카테고리도 함께 연결되어있는 것으로 보아 아동용품으로도 범주를 크게 묶을 수 있을 것으로 판단된다.
