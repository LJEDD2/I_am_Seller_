### í’ˆëª©ë³„ ëˆ„ì  ë§¤ì¶œ

````python
# ì „ì²´ ë§¤ì¶œ
price_sum = df["price"].sum()

# ì¹´í…Œê³ ë¦¬ ê¸°ì¤€ ê°€ê²© í•©ê³„ ê·¸ë£¹í™”
# ê°€ê²© ê¸°ì¤€ ë‚´ë¦¼ì°¨ìˆœ ì •ë ¬
sale_rank = df.groupby("product_category_name_english").agg({"price":"sum"}).sort_values("price", ascending=False)

# ë¹„ìœ¨ ì»¬ëŸ¼ ì¶”ê°€
# ì†Œìˆ˜ 2ì§¸ ìë¦¬ê¹Œì§€ í‘œí˜„
sale_rank["rate"] = round(sale_rank['price'] / price_sum, 2)

# ëˆ„ì  ë¹„ìœ¨
sale_rank["cumulation"] = sale_rank["rate"].cumsum()
sale_rank.head(5)
````

### ê³ ê° ë„ì‹œ ê¸°ì¤€ ì£¼ë¬¸ëŸ‰ ê·¸ë£¹í™”

````python
# ê³ ê° ë„ì‹œ ê¸°ì¤€ ì£¼ë¬¸ëŸ‰ ê·¸ë£¹í™”
# ì£¼ë¬¸ëŸ‰ ê¸°ì¤€ ë‚´ë¦¼ì°¨ìˆœ ì •ë ¬ í›„ top10 í•œì •
city_10 = df.groupby("customer_city").agg({"order_id":"count"}).sort_values("order_id", ascending=False).head(10).index.to_list()

# ì£¼ë¬¸ëŸ‰ì´ ê°€ì¥ ë§ì€ ìƒìœ„ 10ê°œ ë„ì‹œ
city_10
````

````python
# ë„ì‹œ ì£¼ë¬¸ëŸ‰ top10 í•œì •
# í’ˆëª© ë§¤ì¶œ í•©ê³„ top5 í•œì • 
# ëª¨ë‘ í™•ì¸í•˜ë©´ ë„ˆë¬´ ë‚œì¡í•˜ì—¬ í•œì •
top15 = sale_rank.head(15)

# ì¡°ê±´ : (ì¹´í…Œê³ ë¦¬ê°€ ì œí’ˆ top5ì— ìˆê³ ) & (ê³ ê° ë„ì‹œê°€ ë„ì‹œ top10ì— ìˆìœ¼ë©´) == True
condition = (df["product_category_name_english"].map(lambda x: True if x in list(top15.index[:5]) else False)) & (df["customer_city"].map(lambda x: True if x in city_10 else False))

# ì¡°ê±´ì— ë§ëŠ” ë°ì´í„° ì¶”ì¶œ
city = df.loc[condition, :]

# ì œí’ˆ 5ê°œ, ë„ì‹œ 10ê°œê°€ ì˜ ì¶”ì¶œë˜ì—ˆëŠ”ì§€ ê°œìˆ˜ë¡œ í™•ì¸
print(city["product_category_name_english"].nunique(), city["customer_city"].nunique())

# ë°ì´í„° í™•ì¸
city.head()
````

### íŒë§¤ - ì£¼ë¬¸ ì§€ì—­ ê°„ ê±°ë˜ íšŸìˆ˜ ë¹„êµ 

````python
df1 = df_location[['seller_city','customer_city']].copy()

# íŒë§¤ì§€ì—­ê³¼ ì£¼ë¬¸ì§€ì—­ì˜ í•˜ë‚˜ì˜ ìŒ tuple ìƒì„± 
df1['seller_customer_pair'] = df1.apply(lambda x: (x['seller_city'], x['customer_city']), axis = 1)

# value_count ê°œìˆ˜ ì €ì¥ 
seller_single = df1[['seller_city']].apply(pd.value_counts).sum(1) 
customer_single = df1[['customer_city']].apply(pd.value_counts).sum(1)

# ì£¼ë¬¸ì§€ì—­ë³„, íŒë§¤ì§€ì—­ë³„ ê°œìˆ˜
df1 = df1.assign(seller_count = df1['seller_city'].map(seller_single))
df1 = df1.assign(customer_count = df1['seller_city'].map(customer_single))
df1['single_count'] = df1['seller_count'] + df1['customer_count']

pair = df1[['seller_customer_pair']].apply(pd.value_counts).sum(1)
df1 = df1.assign(pair_count = df1['seller_customer_pair'].map(pair))

df1 = df1.drop_duplicates('seller_customer_pair').reset_index(drop = True)
df1[['seller_city','customer_city','seller_customer_pair','pair_count']].head(5)
````

````python
counts = 0
for i in range(df1.shape[0]):
    counts = counts + df1.loc[i,'pair_count']

df2 = df1.drop_duplicates('seller_city')
df2 = df2.sort_values(by = 'single_count', ascending= False).reset_index(drop = True)

print('ì´ ì£¼ë¬¸ ê±´ìˆ˜ : ', counts)
print('ëª¨ë“  seller & customer_citiesì˜ ìœ ë‹ˆí¬í•œ ì¡°í•© ê°œìˆ˜ :', df2.shape[0])
df2.head()
````

````python
x_axis = df2.loc[:9,'seller_city'].to_numpy()
y_axis = x_axis

# íˆíŠ¸ë§µì„ ê·¸ë¦¬ê¸° ìœ„í•œ ë°°ì—´ 
xy = [[0 for i in range(len(x_axis))] for i in range(len(x_axis))]
              
for i in range(df1.shape[0]):
    for ix in range(len(x_axis)):
        if(x_axis[ix] == df1.loc[i,'seller_city']):
            for iy in range(len(y_axis)):
                if(y_axis[iy] == df1.loc[i,'customer_city']):
                    xy[ix][iy] = df1.loc[i,'pair_count']
````

````python
from math import radians , cos, sin, asin, sqrt 
from tqdm import tqdm
def haversine_distance(x1, y1, x2, y2 ):
    x1, y1, x2, y2 = map(radians , [x1,y1,x2,y2])
    d_lng = x2 - x1
    d_lat = y2 - y1
    a = sin(d_lat/2)**2 + cos(y1) * cos(y2) * sin(d_lng/2)**2
    dist = round((2*asin(sqrt(a))*6371*1000)/1000,3)
    return dist 
````

````python
def evaluation_table(df):
    res = df.groupby('product_category_name').agg({"order_id":"count", 
                                                   "product_id":'nunique', 
                                                   "seller_id":'nunique', 
                                                   'price':'sum'})
    res['price_mean'] = round(res['price']/res['seller_id'], 2)
    res['order_mean'] = round(res['order_id']/res['seller_id'], 2)
    res['price_std'] = 0
    res['order_std'] = 0
    res = res.reset_index()

    a = df.groupby(['seller_id', 'product_category_name']).agg({"order_id":"count",
                                                                "product_id":'nunique', 
                                                                'price':'sum'}).reset_index()

    # í‘œì¤€í¸ì°¨
    for i in range(len(res)):
        # price_std
        d = a.loc[a['product_category_name'] == res.iloc[i, 0], 'price'] - res.iloc[i, 5]
        s = round((sum(d**2)/len(d))**1/2, 2)
        res.iloc[i, 7] = s

        # order_std
        d = a.loc[a['product_category_name'] == res.iloc[i, 0], 'order_id'] - res.iloc[i, 6]
        s = round((sum(d**2)/len(d))**1/2, 2)
        res.iloc[i, 8] = s
    
    return res
````

````python
def market_cluster(df):
    if len(df) < 4:
        return 0
    # íŒë§¤ìì˜ ë“±ê¸‰ì„ êµ°ì§‘í™” 
    scaler = StandardScaler()

    # K-Means Clustering 
    model = KMeans(n_clusters=4, random_state=42)

    pipeline = make_pipeline(scaler, model)
    pipeline.fit(df[['order_id', 'price']])

    market = df[['order_id', 'price']].copy()
    market["cluster"] = pipeline.predict(df[['order_id', 'price']])
    market['product_category_name'] = df['product_category_name']  
    return market

def cluster_tier(df, zero, one, two, three):
    df['cluster'] = df['cluster'].replace(0, zero)
    df['cluster'] = df['cluster'].replace(1, one)
    df['cluster'] = df['cluster'].replace(2, two)
    df['cluster'] = df['cluster'].replace(3, three)
    
    plt.figure(figsize=(6, 6))
    sns.scatterplot(data=df, x='order_id', y='price', hue='cluster', palette="Set1")

    return df
````
