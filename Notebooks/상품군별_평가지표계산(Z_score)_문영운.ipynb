```python
top8_cat = np.array(category_21[final_category][13:].sum().index)
top8_cluster = category_21[final_category][13:].sum().values
top8_z_total= z_total[13:].sum().values
```

```python
top8_rfm = pd.read_csv(raw_path + 'Top8_RFM.csv')
top8_rfm['RFM_total'] = round(top8_rfm['RFM_total'], 2)
top8_rfm = top8_rfm.replace('house_wares', 'housewares')
top8_rfm
```

```python
top8_evaluation = pd.merge(top8_etc, top8_rfm, on='category').sort_values('category').reset_index(drop=True)
top8_evaluation
```

```python
def standardization(df, col):
    m = (df[col].describe()).loc['mean']
    s = (df[col].describe()).loc['std']
    z = (df[col] - m)/s
    return z
```

```python
top8_evaluation['Z'] = standardization(top8_evaluation, 'cluster') + standardization(top8_evaluation, 'Z_total') +  standardization(top8_evaluation, 'RFM_total')
top8_evaluation.sort_values('Z', ascending=False).reset_index(drop=True)
```

```python
top8_evaluation['score_2'] = round(standardization(top8_evaluation, 'cluster'), 2) + round(standardization(top8_evaluation, 'Z_total'), 2) +  (top8_evaluation['RFM_total'])
top8_evaluation.sort_values('score_2', ascending=False).reset_index(drop=True)
```
